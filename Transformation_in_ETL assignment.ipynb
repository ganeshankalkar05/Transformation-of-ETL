
#Transformation in ETL

Question 1 : Define Data Transformation in ETL and explain why it is important?
  ->Data Transformation is the "T" in the ETL (Extract, Transform, Load) process. It is the stage where raw data extracted from various sources is cleaned, structured, and enriched to fit the operational needs of the target database or data warehouse.

Why it is important:
Raw data is rarely ready for analysis right out of the box. It often comes in different formats, contains errors, or includes irrelevant information. Transformation is crucial because it acts as a quality control checkpoint. It ensures that the final data is consistent, accurate, and properly formatted, which directly determines the reliability of any business intelligence, reporting, or machine learning models built on top of it.

Question 2 : List any four common activities involved in Data Cleaning?

   ->Handling Missing Values: Filling in blank data points using statistical methods (imputation) or removing incomplete records entirely.

Deduplication: Identifying and deleting identical, repeated rows of data to prevent skewed analysis.

Fixing Structural Errors: Correcting typos, standardizing capitalization, and ensuring consistent naming conventions across datasets.

Filtering Outliers: Identifying extreme values that don't fit the dataset's normal distribution and deciding whether to remove or adjust them.

Question 3 : What is the difference between Normalization and Standardization?
   ->Normalization (Min-Max Scaling): Rescales the data so that all values fit within a strict, specific boundary, typically between 0 and 1. It is highly influenced by the maximum and minimum values in the dataset.
   ->Standardization (Z-Score Scaling): Centers the data around a mean of 0 with a standard deviation of 1. It does not bound the data to a specific range, making it much more practical when your data follows a normal (bell curve) distribution.

Question 4 : A dataset has missing values in the “Age” column. Suggest two techniques to handle this and explain when thwy should be used?
  ->Median Imputation: You replace the missing values with the middle value of all the known ages in the dataset.

When to use it: This is ideal if your dataset contains outliers (e.g., a few 95-year-olds in a dataset mostly containing college students), because the median is not easily skewed by extreme highs or lows.

Row Deletion (Dropping): You simply delete the entire row of data for the person missing their age.

When to use it: This should only be used if the missing values represent a very tiny fraction of a massive dataset (e.g., less than 2%) and dropping them will not introduce bias or cost you valuable information from other columns.

Question 6 : What is One-Hot Encoding? Give an example with the categories: “Red, Blue, Green”?
  
  ->One-Hot Encoding is a process that converts categorical data (words or text) into a binary matrix (0s and 1s) so that machine learning algorithms can process it. It prevents the algorithm from assuming that categories have a numerical ranking or order.

Example with Red, Blue, Green:
Instead of a single "Color" column, One-Hot Encoding creates a new, separate column for every possible category. If a row is "Red", it gets a 1 in the Red column and a 0 in the others.
Original Color,Is_Red,Is_Blue,Is_Green
Red,              1       0      0
Blue              0       1      0
Green             0       0      1

Question 7 : Explain the difference between Data Integration and Data Mapping in ETL?

   ->Data Integration is the broad, overarching strategy of combining data from completely different systems (like a CRM, a billing platform, and a marketing tool) into one unified, cohesive view.

Data Mapping is a specific, tactical step inside the integration process. It is the act of defining exactly how a specific field from Source A translates to a field in Target B. For example, explicitly defining that the client_id column in your sales database should flow into the CustomerNumber column in your central data warehouse.

Question 8 : Explain why Z-score Standardization is preferred over Min-Max Scaling when outliers exist?

   ->Z-score Standardization is preferred when outliers exist because Min-Max Scaling is entirely dependent on the absolute minimum and maximum values in the dataset.

If you have an extreme outlier (e.g., most salaries are between 50,000 and 100,000, but one is 10,000,000), Min-Max scaling will use that 10,000,000 as its upper bound. This will compress all your normal, everyday salaries into a tiny, indistinguishable fraction near 0. Z-score scaling uses the mean and standard deviation, which spreads the core data out more naturally and preserves the relative distances between normal data points, leaving the outlier as a recognizable extreme value rather than letting it ruin the scale for everything else.
